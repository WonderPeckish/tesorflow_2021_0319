```python
X = tf.placeholder(tf.float32, [None, 3])
x_data = [ [1,2,3],[4,5,6] ]
W = tf.Variable(tf.random_normal([3,2]))
b = tf.Variable(tf.random_normal([2,1]))
result = tf.matmul(X,W)+b

sess = tf.Session()
sess.run(tf.global_variables_initializer())

print(sess.run(W))
print(sess.run(b))
print(x_data)

result_temp = sess.run(result, feed_dict={X:x_data})
print(result_temp)
sess.close()
```

```python
x_data = [1,2,3]
y_data = [1,2,3]

W=tf.Variable(tf.random_uniform([1], -0.1, 1.0))
b=tf.Variable(tf.random_uniform([1], -0.1, 1.0))

X=tf.placeholder(tf.float32, name="X")
Y=tf.placeholder(tf.float32, name="Y")

hypothesis = W*X+b
cost = tf.reduce_mean(tf.square(hypothesis-Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train_op = optimizer.minimize(cost)

with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(100):
    _, cost_value = sess.run([train_op, cost], feed_dict={X:x_data, Y:y_data})
    print(step, cost_value, sess.run(W), sess.run(b))

  print("===test===")
  print("x : 5, Y:", sess.run(hypothesis, feed_dict={X:5}))
  print("x : 2.5, Y:", sess.run(hypothesis, feed_dict={X:2.5}))
```
